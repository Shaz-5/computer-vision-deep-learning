```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

# Image Stitching

```python
# Loading images
image1 = cv2.imread('./flower1.jpeg')
plt.figure(figsize=(8,8))
plt.imshow(cv2.cvtColor(image1, cv2.COLOR_BGR2RGB))
```

```python
image2 = cv2.imread('./flower2.jpeg')
plt.figure(figsize=(8,8))
plt.imshow(cv2.cvtColor(image2, cv2.COLOR_BGR2RGB))
```

```python
# Stitcher object
stitcher = cv2.Stitcher.create()

# Stitch the images
(status, stitched) = stitcher.stitch([image1, image2])

plt.figure(figsize=(8,8))

# If the stitching was successful
if status == cv2.Stitcher_OK:
    plt.imshow(cv2.cvtColor(stitched, cv2.COLOR_BGR2RGB))
else:
    print("Image stitching failed.")
```

# Image Segmentation

```python
image = stitched

gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Canny edge detection
edges = cv2.Canny(gray, 0, 200)

# Contours 
contours, hierarchy = cv2.findContours(edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Draw the contours on the original image
segmented_image = cv2.drawContours(image, contours, -1, (0, 255, 0), 6)
```

```python
plt.figure(figsize=(10,10))
plt.imshow(cv2.cvtColor(segmented_image, cv2.COLOR_BGR2RGB))
```
